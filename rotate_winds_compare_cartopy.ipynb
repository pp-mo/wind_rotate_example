```python
import matplotlib.pyplot as plt
import numpy as np
import iris
import cartopy
import cartopy.crs as ccrs
```

```python
# A rotated system with its latlon=0,0 at real latlon=60,0
cs_pc = iris.coord_systems.GeogCS(iris.fileformats.pp.EARTH_RADIUS)
pole_lon, pole_lat = 180, 30
cs_rp = iris.coord_systems.RotatedGeogCS(grid_north_pole_latitude=pole_lat, grid_north_pole_longitude=pole_lon, ellipsoid=cs_pc)
crs_rp = cs_rp.as_cartopy_projection()
crs_pc = ccrs.PlateCarree()

# Prove that...
plt.figure(figsize=(12,8))
ax = plt.axes(projection=crs_pc)
ax.coastlines()
ax.gridlines(crs=crs_pc)
ax.gridlines(crs=crs_rp, color='green', linestyle='-')
plt.title('Normal map (rotated grid in green).')
plt.show()

plt.figure(figsize=(12,8))
ax = plt.axes(projection=crs_rp)
ax.coastlines()
ax.gridlines(crs=crs_pc)
ax.gridlines(crs=crs_rp, color='green', linestyle='-')
plt.title('Rotated map.')
plt.show()
```

```python
# Show the local region around its (0,0)
plt.figure(figsize=(12,8))
ax = plt.axes(projection=crs_rp)
ax.coastlines()
ax.gridlines(crs=crs_pc)
ax.gridlines(crs=crs_rp, color='green', linestyle='-')
ax.set_extent((-40, 40, -40, 40), crs=crs_rp)
plt.title('Region around 0,0 in rotated system (true-lat,lon=60,0).')
plt.show()
```

```python
# Define a cube with U,V=1,1 at rotated_lat,rotated_lon=0,0
xx = np.array([0.0])
yy = np.array([0.0])

uu = np.array([1.0])
vv = np.array([1.0])

x_coord = iris.coords.DimCoord(xx, standard_name='grid_longitude', units='degrees',
                               coord_system=cs_rp)
y_coord = iris.coords.DimCoord(yy, standard_name='grid_latitude', units='degrees',
                               coord_system=cs_rp)

cube_u = iris.cube.Cube([uu], long_name='u', units='ms-1')
cube_v = iris.cube.Cube([uu], long_name='u', units='ms-1')
for cube in (cube_u, cube_v):
    cube.add_dim_coord(y_coord, 0)
    cube.add_dim_coord(x_coord, 1)
print cube_u
```

```python
# Rotate to "true" lat-lon with Iris method
xt, yt = iris.analysis.cartography.unrotate_pole(xx, yy, pole_lon, pole_lat)
cube_ut, cube_vt = iris.analysis.cartography.rotate_winds(cube_u, cube_v, cs_pc)
print 'From Iris:'
print 'X=', xt
print 'Y=', yt
print 'U=', cube_ut.data
print 'V=', cube_vt.data
```

```python
# Do same with cartopy
p3d = crs_pc.transform_points(crs_rp, xx, yy)
xt = p3d[..., 0]
yt = p3d[..., 1]
ut, vt = crs_pc.transform_vectors(crs_rp, xx, yy, uu, vv)
print 'From Cartopy:'
print 'X=', xt
print 'Y=', yt
print 'U=', ut
print 'V=', vt
```
